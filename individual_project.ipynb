```r
#run this code to read data 
library(tidyverse)

players_data <- read.csv("data/players.csv") |>
    select(name, gender, Age, played_hours, experience, subscribe, -hashedEmail, -gender) |>
    rename(age = Age) |>
    arrange(age)

players_data 
```

```r
#Compute mean value for each qualitative variable
summary_mean <- players_data |>
    summarize(mean_age = mean(age, na.rm = TRUE),
              mean_played_hours = mean(played_hours, na.rm = TRUE)) 

summary_mean
```

```r
#Report mean values in table format 
summary_table <- summary_mean |>
    pivot_longer(cols = everything(), 
                 names_to = "variable", 
                 values_to = "mean")
summary_table 
```
